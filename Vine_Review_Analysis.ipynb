```python
#import dependencies

import pandas as pd
import os
```

```python
#read in data file
vines_df = pd.read_csv("vine_reviews.csv")
vines_df
```

```python
#Filter the data to create a new DataFrame where total_votes >=20
total_votes_20 = vines_df[(vines_df["total_votes"] >= 20)]
total_votes_20
```

```python
#Retrieve all the rows where the number of helpful_votes divided by the total_votes is equal to or greater than 50%
helpful50_df = total_votes_20[(total_votes_20["helpful_votes"]/total_votes_20["total_votes"]>=0.5)]

helpful50_df
```

```python
#Filter the >50% helpful reviews for vine =Y, a review that was part of the vine program
helpful_paid = helpful50_df[(helpful50_df["vine"]=="Y")]
helpful_paid
```

```python
#Filter the >50% helpful reviews for vine =Y, a review that was not part of the vine program
helpful_unpaid = helpful50_df[(helpful50_df["vine"]=="N")]
helpful_unpaid
```

```python
# Calculate the count of the helpful reviews that were part of the vine program
helpful_paid_count = helpful_paid["vine"].count()

#Caclulate the count of the paid helpful reviews that were 5 stars
helpful_paid_5stars = helpful_paid[(helpful_paid["star_rating"])==5].count()
helpful_paid_5stars=helpful_paid_5stars["review_id"]

#Calculate the percentage of paid helpful reviews that were 5 stars
helpful_paid_5stars_pct = helpful_paid_5stars/float(helpful_paid_count)
helpful_paid_5stars_pct
```

```python
# Calculate the count of the helpful reviews that were not part of the vine program
helpful_unpaid_count = helpful_unpaid["vine"].count()

#Caclulate the count of the unpaid helpful reviews that were 5 stars
helpful_unpaid_5stars = helpful_unpaid[(helpful_unpaid["star_rating"])==5].count()
helpful_unpaid_5stars=helpful_unpaid_5stars["review_id"]

#Calculate the percentage of unpaid helpful reviews that were 5 stars
helpful_unpaid_5stars_pct = helpful_unpaid_5stars/float(helpful_unpaid_count)
helpful_unpaid_5stars_pct
```
